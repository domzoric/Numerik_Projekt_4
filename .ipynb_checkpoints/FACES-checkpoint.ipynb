# Eigenfaces

## <font style="color:rgb(50,120,229)">Was sind Eigenfaces?</font>

Eigenfaces sind Bilder, die zu einem Durchschnittsgesicht hinzugefügt werden können, um neue Gesichter zu erzeugen. Mathematisch können wir dies wie folgt ausdrücken:

$$
F = F_{m} + \sum \limits _{i=1} ^{n} \alpha_{i}F_{i}
$$

wobei:

- $F$ ein neues Gesicht ist.
- $F_{m}$ das Mittel- oder Durchschnittsgesicht ist.
- $F_{i}$ ein Eigenface ist.
- $\alpha_{i}$ Skalare Multiplikatoren sind, die wir wählen können, um neue Gesichter zu erstellen. Sie können positiv oder negativ sein.

Eigenfaces werden berechnet, indem die Hauptkomponenten (Principal Components) des Datensatzes von Gesichtsbildern geschätzt werden. Sie werden in Anwendungen wie Gesichtserkennung und Erkennung von Gesichtspunkten verwendet.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from ipywidgets import interactive, fixed
import ipywidgets as widgets
from IPython.display import display
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import pandas as pd
```

Überlegung: Datenkompression durch Dimesnionsreduktion

```python
original_size = faces.nbytes / (1024 ** 2)  # Größe in Megabyte
original_size
```

Überlegung Anzahlhauptkomponeneten mit dem Silhoute Score berechenen

```python
# Liste der Anzahl der Hauptkomponenten, die getestet werden sollen
n_components_list = range(2, 101, 5)

# Speicher für die Silhouette-Koeffizienten
silhouette_scores = []

# K-Means Konfiguration
n_clusters = 40  # Anzahl der Cluster

for n_components in n_components_list:
    pca = PCA(n_components=n_components)
    faces_pca = pca.fit_transform(faces)
    
    # K-Means auf den PCA-transformierten Daten ausführen
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(faces_pca)
    
    # Silhouette-Koeffizient berechnen
    silhouette_avg = silhouette_score(faces_pca, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Beste Anzahl der Hauptkomponenten bestimmen
best_n_components = n_components_list[np.argmax(silhouette_scores)]

print(f'Die optimale Anzahl der Hauptkomponenten basierend auf dem höchsten Silhouette-Koeffizienten ist: {best_n_components}')

# Silhouette-Koeffizienten plotten
plt.figure(figsize=(8, 6))
plt.plot(n_components_list, silhouette_scores, marker='o', linestyle='--')
plt.axvline(x=best_n_components, color='r', linestyle='--', label=f'Optimum bei {best_n_components} Komponenten')
plt.xlabel('Anzahl der Hauptkomponenten')
plt.ylabel('Silhouette-Koeffizient')
plt.title('Silhouette-Koeffizient in Abhängigkeit von der Anzahl der Hauptkomponenten')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PCA auf 3 Dimensionen für die Visualisierung
pca_3d = PCA(n_components=3)
faces_pca_3d = pca_3d.fit_transform(faces)

# K-Means Clustering mit 40 Clustern (entspricht den 40 Klassen)
n_clusters = 40
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(faces_pca_3d)
cluster_centers = kmeans.cluster_centers_

# Berechne die durchschnittliche Distanz der Clusterzentren
distances = np.zeros((n_clusters, n_clusters))
for i in range(n_clusters):
    for j in range(i + 1, n_clusters):
        distances[i, j] = np.linalg.norm(cluster_centers[i] - cluster_centers[j])
        distances[j, i] = distances[i, j]

avg_distances = np.mean(distances, axis=1)
top_3_clusters = np.argsort(avg_distances)[-3:]

# Funktion zum Plotten der Cluster in 3D
def plot_3d(elev, azim):
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    for cluster in top_3_clusters:
        mask = cluster_labels == cluster
        ax.scatter(faces_pca_3d[mask, 0], faces_pca_3d[mask, 1], faces_pca_3d[mask, 2], label=f'Cluster {cluster}')
    
    ax.set_xlabel('Erste Hauptkomponente')
    ax.set_ylabel('Zweite Hauptkomponente')
    ax.set_zlabel('Dritte Hauptkomponente')
    ax.set_title('K-Means Cluster auf 3D PCA-reduzierten Daten')
    ax.view_init(elev=elev, azim=azim)
    ax.legend()
    plt.show()

# Interaktives Widget erstellen
elev_slider = widgets.FloatSlider(min=0, max=90, step=1, value=30, description='Elevation')
azim_slider = widgets.FloatSlider(min=-180, max=180, step=1, value=30, description='Azimuth')
interactive_plot = interactive(plot_3d, elev=elev_slider, azim=azim_slider)

# Anzeige des interaktiven Widgets
display(interactive_plot)
```

```python
# Zeige repräsentative Bilder der ausgewählten Cluster
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, cluster in zip(axes, top_3_clusters):
    mask = cluster_labels == cluster
    cluster_faces_pca = faces_pca_3d[mask]
    cluster_center = cluster_centers[cluster]
    
    # Find the closest face in the PCA-reduced space
    closest_face_idx = np.argmin(np.linalg.norm(cluster_faces_pca - cluster_center, axis=1))
    
    # Get the corresponding face in the original space
    closest_face = dataset.images[mask][closest_face_idx]
    ax.imshow(closest_face, cmap='gray')
    ax.set_title(f'Cluster {cluster}')
    ax.axis('off')

plt.show()
```

```python
# Zeige drei repräsentative Bilder der ausgewählten Cluster nach Rücktransformation
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

for i, cluster in enumerate(top_3_clusters):
    mask = cluster_labels == cluster
    cluster_faces_pca = faces_pca_3d[mask]
    cluster_center = cluster_centers[cluster]
    
    # Find the three closest faces in the PCA-reduced space
    distances = np.linalg.norm(cluster_faces_pca - cluster_center, axis=1)
    closest_faces_indices = np.argsort(distances)[:3]
    
    for j, idx in enumerate(closest_faces_indices):
        closest_face_pca = pca_3d.inverse_transform(cluster_faces_pca[idx])
        closest_face = closest_face_pca.reshape((64, 64))  # reshape to original dimensions if necessary
        axes[i, j].imshow(closest_face, cmap='gray')
        if j == 0:
            axes[i, j].set_title(f'Cluster {cluster}')
        axes[i, j].axis('off')

plt.show()
```

```python
pca_1d = PCA(n_components=1)
faces_pca_1d = pca_1d.fit_transform(faces)
first_component = pca.components_[0]
eigenface = first_component.reshape(64, 64)

plt.figure(figsize=(6, 6))
plt.imshow(eigenface, cmap='gray')
plt.title('Erstes Eigenface (erste Hauptkomponente)')
plt.axis('off')
plt.show()
```

```python
# PCA ohne Angabe der Anzahl der Komponenten
pca = PCA()
pca.fit(faces)

# Kumulierte erklärte Varianz plotten
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.xlabel('Anzahl der Hauptkomponenten')
plt.ylabel('Kumulierte erklärte Varianz')
plt.title('Kumulierte erklärte Varianz in Abhängigkeit der Anzahl der Hauptkomponenten')
plt.grid(True)
plt.show()
```

```python
# Erklärte Varianz und kumulierte erklärte Varianz berechnen
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)

# Anzahl der Komponenten für 95% der Varianz
threshold = 0.95
num_components = np.argmax(cumulative_explained_variance >= threshold) + 1

print(f'Die Anzahl der Hauptkomponenten, die 95% der Varianz erklären, ist: {num_components}')

# Plot mit dem identifizierten Punkt für 95% der Varianz
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.axvline(x=num_components, color='r', linestyle='--', label=f'95% Varianz bei {num_components} Komponenten')
plt.xlabel('Anzahl der Hauptkomponenten')
plt.ylabel('Kumulierte erklärte Varianz')
plt.title('Kumulierte erklärte Varianz mit 95% Schwellenwert')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PCA auf Bildermatrix anwenden
n_components = 123  # Anzahl der Hauptkomponenten
pca = PCA(n_components=n_components, whiten=True)
faces_pca = pca.fit_transform(faces)
compressed_size = (faces_pca.nbytes + pca.components_.nbytes) / (1024 ** 2)  # Größe in Megabyte
compressed_size
```

```python
# Ein Bild komprimieren und rekonstruieren
face_idx = 0  # Index des zu komprimierenden Bildes
original_face = faces[face_idx]
compressed_face = pca.inverse_transform(faces_pca[face_idx].reshape(1, -1)).reshape(-1)
```

```python
# Original und rekonstruierte Bilder anzeigen
fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].imshow(original_face.reshape((64, 64)), cmap='gray')
ax[0].set_title('Originalbild')

ax[1].imshow(compressed_face.reshape((64, 64)), cmap='gray')
ax[1].set_title('Rekonstruiertes Bild')

plt.show()

# Überprüfe die Dimensionen
print("Original shape:", original_face.shape)
print("Compressed shape:", compressed_face.shape)
```

```python
# Einige der Hauptkomponenten visualisieren
fig, axes = plt.subplots(2, 5, figsize=(15, 6), subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(64, 64), cmap='gray')
    ax.set_title(f'PC {i+1}')

plt.show()
```

### Part II

```python
# YALE-Face-Datenbank laden (hier verwenden wir Olivetti Faces als Beispiel)
dataset = fetch_olivetti_faces(shuffle=True, random_state=42)
faces = dataset.data
labels = dataset.target
n_samples, n_features = faces.shape
```

```python
# Daten in Trainings- und Testdaten aufteilen
X_train, X_test, y_train, y_test = train_test_split(faces, labels, test_size=0.3, random_state=42)
```

```python
# PCA auf die Trainingsdaten anwenden
n_components = 123  # Anzahl der Hauptkomponenten
pca = PCA(n_components=n_components, whiten=True)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Klassifikation im Hauptachsenraum
def classify_face(face_pca, X_train_pca, y_train):
    distances = np.linalg.norm(X_train_pca - face_pca, axis=1)
    return y_train[np.argmin(distances)]
```

```python
# Vorhersagen für die Testdaten treffen
y_pred = [classify_face(face, X_train_pca, y_train) for face in X_test_pca]
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Genauigkeit der Gesichtserkennung: {accuracy:.2f}')
```

```python
def plot_results(X_test, y_test, y_pred, images_per_row=5):
    fig, axes = plt.subplots(images_per_row, images_per_row, figsize=(10, 10))
    fig.suptitle('Tatsächliche vs. vorhergesagte Labels', size=16)
    for i in range(images_per_row):
        for j in range(images_per_row):
            idx = i * images_per_row + j
            if idx >= len(y_test):
                break
            ax = axes[i, j]
            ax.imshow(X_test[idx].reshape((64, 64)), cmap='gray')
            ax.set_title(f'Tatsächlich: {y_test[idx]}\nVorhergesagt: {y_pred[idx]}')
            ax.axis('off')
    plt.show()

# Plot some results
plot_results(X_test, y_test, y_pred, images_per_row=5)
```

```python
# PCA auf die Trainingsdaten anwenden (Reduktion auf 3 Hauptkomponenten)
n_components = 3
pca_3d = PCA(n_components=n_components, whiten=True)
X_train_pca_3d = pca_3d.fit_transform(X_train)
X_test_pca_3d = pca_3d.transform(X_test)

# K-Means Clustering auf die PCA-reduzierten Daten anwenden (für 3 Cluster)
n_clusters = 3
kmeans_3d = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_3d.fit(X_train_pca_3d)
labels_train_3d = kmeans_3d.labels_
cluster_centers_3d = kmeans_3d.cluster_centers_

# 3D-Plot der Cluster
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Cluster in verschiedenen Farben plotten
for cluster in range(n_clusters):
    mask = labels_train_3d == cluster
    ax.scatter(X_train_pca_3d[mask, 0], X_train_pca_3d[mask, 1], X_train_pca_3d[mask, 2], label=f'Cluster {cluster}')

# Clusterzentren plotten
ax.scatter(cluster_centers_3d[:, 0], cluster_centers_3d[:, 1], cluster_centers_3d[:, 2], s=300, c='red', marker='X', label='Cluster Zentren')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('K-Means Cluster auf 3D PCA-reduzierten Daten')
ax.legend()
plt.show()
```

```python
# Anzahl der Klassen beschränken (z.B. die 10 häufigsten Klassen)
most_common_classes = np.argsort(np.bincount(y_test))[-10:]

# Mask für häufigste Klassen erstellen
mask = np.isin(y_test, most_common_classes)

# Gefilterte Labels und Vorhersagen für häufigste Klassen
y_test_filtered = np.array(y_test)[mask]
y_pred_filtered = np.array(y_pred)[mask]

# Gefilterte Confusion Matrix berechnen
filtered_cm = confusion_matrix(y_test_filtered, y_pred_filtered, labels=most_common_classes)

# Gefilterte Confusion Matrix anzeigen
plt.figure(figsize=(10, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=filtered_cm, display_labels=most_common_classes)
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title('Confusion Matrix der häufigsten Klassen')
plt.show()
```

```python
# Ein Testbild und das zugehörige Originalbild anzeigen
test_idx = 0  # Index des Testbildes
test_face = X_test[test_idx]
predicted_label = classify_face(X_test_pca[test_idx], X_train_pca, y_train)
true_label = y_test[test_idx]

fig, ax = plt.subplots(1, 2, figsize=(8, 4))

ax[0].imshow(test_face.reshape((64, 64)), cmap='gray')
ax[0].set_title(f'Original (Klasse: {true_label})')

# Ein Trainingsbild aus der vorhergesagten Klasse anzeigen
predicted_face = X_train[y_train == predicted_label][0]

ax[1].imshow(predicted_face.reshape((64, 64)), cmap='gray')
ax[1].set_title(f'Vorhergesagt (Klasse: {predicted_label})')

plt.show()
```
