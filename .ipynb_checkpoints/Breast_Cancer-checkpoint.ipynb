## Warum verwenden wir einen Standard Scaler fÃ¼r PCA?

- **PCA ist empfindlich gegenÃ¼ber Skalierung:**  
  PCA basiert auf der Kovarianzmatrix der Daten, die die Varianz der Daten in verschiedenen Dimensionen misst. Variablen mit grÃ¶ÃŸeren Wertebereichen kÃ¶nnen die Ergebnisse dominieren und die Analyse verzerren.

- **Beispiel:**  
  Stellen wir uns zwei Merkmale vor:  
  - **Feature A:** Variiert von 0 bis 1 (z.B. eine normalisierte Variable)
  - **Feature B:** Variiert von 0 bis 10.000 (z.B. Einkommen in Euro)
  
  Ohne Skalierung wÃ¼rde **Feature B** den grÃ¶ÃŸten Teil der Varianz erklÃ¤ren, nur aufgrund seines grÃ¶ÃŸeren Wertebereichs, auch wenn **Feature A** ebenfalls wichtig ist.

- **Standard Scaler:**  
  Skaliert die Daten so, dass jede Variable einen Mittelwert von 0 und eine Standardabweichung von 1 hat. Dadurch werden alle Merkmale gleichgewichtet und die PCA kann die wahre Struktur der Daten erkennen.

- **Fazit:**  
  Die Standardisierung der Daten vor der DurchfÃ¼hrung der PCA stellt sicher, dass alle Merkmale fair berÃ¼cksichtigt werden, unabhÃ¤ngig von ihren ursprÃ¼nglichen Einheiten oder Wertebereichen.


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

# PCA-Transformation

Die Hauptkomponentenanalyse (PCA) ist eine Technik zur Dimensionsreduktion, die die Kovarianzmatrix der Daten verwendet. Die Schritte zur DurchfÃ¼hrung einer PCA-Transformation sind wie folgt:

1. **Zentriere die Daten**:
   $$ X_{\text{zentriert}} = X - \mu $$

   **Relevanz**: Die Zentrierung der Daten (Subtraktion des Mittelwerts $\mu$ von jedem Merkmal) ist entscheidend, da die PCA die Daten um den Ursprung zentriert. Dies stellt sicher, dass die Hauptkomponenten (die neuen Achsen) die maximale Varianz der Daten erfassen und dass die Analyse nicht durch unterschiedliche Skalierungen der Merkmale verzerrt wird.


```python
# 1. Zentrieren des Datensatzes
X_mean = np.mean(scaled_data, axis=0) #Mittelwert fÃ¼r jedes Feature berechnet 
X_centered = scaled_data - X_mean
X_centered
```

2. **Berechne die Kovarianzmatrix**:
   $$ \Sigma = \frac{1}{n-1} X_{\text{zentriert}}^T X_{\text{zentriert}} $$

**Relevanz**: Die Kovarianzmatrix $\Sigma$ fasst die Varianz und die Korrelation zwischen den verschiedenen Merkmalen zusammen. Sie zeigt, wie die Merkmale gemeinsam variieren. Eigenvektoren der Kovarianzmatrix geben die Richtungen der maximalen Varianz (d.h. die neuen Achsen) an, wÃ¤hrend die Eigenwerte die StÃ¤rke dieser Varianz in den jeweiligen Richtungen ausdrÃ¼cken.

**$ \Sigma $ hat folgende Eigeschaften:**

- Die Komponente $\sigma^2_{ij}$ ist gleich der Kovarianz der Feature $x_i$ und $x_j$
- Auf der diagonalen stehen die Varianzen der einzelnen Feature
- $ \Sigma $ ist symmetrisch und die Eigenvektoren bilden eine Orthononormalbasis
- Sei $ \lambda_1 $ der grÃ¶ÃŸte Eigenwert von $ \Sigma $, dann hat der Datensatz in Richtung $ v_1 $ die grÃ¶ÃŸte Varianz

```python
# 2. Berechnung der Kovarianzmatrix
cov_matrix = np.cov(X_centered, rowvar=False)
cov_matrix.shape
```

```python
cov_matrix = np.cov(X_centered, rowvar=False)

plt.figure(figsize=(10, 8))
sns.heatmap(cov_matrix, cmap='coolwarm', annot=False)
plt.title('Covariance Matrix')
plt.show()
```

3. **FÃ¼hre die Eigenwertzerlegung der Kovarianzmatrix durch**:
   $$ \Sigma v = \lambda v $$

   **Relevanz**: Die Eigenwertzerlegung zerlegt die Kovarianzmatrix in ihre Eigenwerte ($\lambda$) und Eigenvektoren ($v$). Die Eigenvektoren reprÃ¤sentieren die Hauptkomponenten, d.h. die neuen Achsen des Datenraums, entlang derer die Varianz maximiert wird. Die Eigenwerte geben an, wie viel Varianz jede dieser neuen Achsen erklÃ¤rt.

```python
# 3. Eigenwertzerlegung der Kovarianzmatrix
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
eigenvalues
```

```python
# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvalues in descending order
sorted_eigenvalues = np.sort(eigenvalues)[::-1]

# Plotting the eigenvalues
plt.figure(figsize=(8, 6))
plt.plot(sorted_eigenvalues, 'o-')
plt.title('Eigenvalues of the Covariance Matrix')
plt.xlabel('Index of Eigenvalue')
plt.ylabel('Eigenvalue')
plt.grid(True)
plt.show()
```

4. **Sortiere die Eigenvektoren und wÃ¤hle die grÃ¶ÃŸten aus**:
   WÃ¤hle die $k$ grÃ¶ÃŸten Eigenvektoren aus, um die Transformationsmatrix ($W$) zu bilden.

**Relevanz**: Durch das Sortieren der Eigenvektoren nach ihren entsprechenden Eigenwerten (in absteigender Reihenfolge) und das AuswÃ¤hlen der $k$ grÃ¶ÃŸten, behalten wir die wichtigsten Hauptkomponenten bei. Diese Hauptkomponenten erklÃ¤ren den grÃ¶ÃŸten Teil der Varianz in den Daten, wodurch die DimensionalitÃ¤t reduziert wird, wÃ¤hrend die wichtigsten Informationen erhalten bleiben.
   

Sei $W$ eine Matrix, deren Spalten die orthonormalen Eigenvektoren von $\Sigma$ hat, damit gilt $W^-1 = W^T$ 



```python
# Die Eigenwerte und Eigenvektoren sortieren
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]
first_two_eigenvectors = sorted_eigenvectors[:, :2]
first_two_eigenvectors
first_two_eigenvectors
```

```python
scaling_factor = 5 
scaled_eigenvectors = first_two_eigenvectors * scaling_factor
plt.figure(figsize=(8, 8))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=cancer.target, cmap='viridis', edgecolor='k', s=50)
plt.xlabel(cancer.feature_names[0])
plt.ylabel(cancer.feature_names[1])
plt.title('Originaldaten mit Eigenvektoren')
origin = np.zeros(2)  
for i in range(2):
    plt.quiver(*origin, 
               scaled_eigenvectors[0, i], scaled_eigenvectors[1, i], 
               scale=1, scale_units='xy', angles='xy', 
               color=['r', 'b'][i], width=0.005, label=f'PC{i+1}')
plt.legend()
plt.grid(True)
plt.show()
```

5. **Transformiere die Daten in den neuen Raum**:
   $$ X_{\text{PCA}} = X_{\text{zentriert}} W $$

   **Relevanz**: In diesem Schritt werden die ursprÃ¼nglichen Daten in den Raum der Hauptkomponenten transformiert. Die neue Darstellung $X_{\text{PCA}}$ hat reduzierte Dimensionen, enthÃ¤lt aber die maximale Varianz der Originaldaten. Dies ermÃ¶glicht eine effizientere und aussagekrÃ¤ftigere Analyse der Daten, da irrelevante Informationen und Rauschen reduziert werden.

```python
X_pca_manual = np.dot(X_centered, sorted_eigenvectors)
X_pca_manual
```

```python
# Reduzierung auf die ersten 2 Hauptkomponenten
X_pca_2d = X_pca_manual[:, :2]

# Erste 5 Zeilen der 2D projizierten Daten anzeigen
X_pca_2d[:5]
```

```python
# Create a figure
plt.figure(figsize=(8, 6))

# Scatter plot with colors based on the 'target' column
scatter = plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=cancer['target'], cmap='coolwarm', alpha=0.7)

# Adding labels to the axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')

# Adding a legen
plt.legend(handles=scatter.legend_elements()[0], labels=['No Cancer', 'Cancer'])

# Show the plot
plt.show()
```


Zusammengefasst ist die PCA-Transformation:
   $$ X_{\text{PCA}} = (X - \mu) W $$

### RÃ¼cktransformation bei der PCA

Die RÃ¼cktransformation bei der PCA dient dazu, die reduzierten Daten wieder in den ursprÃ¼nglichen Merkmalsraum zurÃ¼ckzufÃ¼hren. Dies kann nÃ¼tzlich sein, um zu verstehen, wie gut die reduzierten Komponenten die ursprÃ¼nglichen Daten reprÃ¤sentieren.

#### 1. RÃ¼cktransformation: Mathematische Herleitung

Nachdem die PCA durchgefÃ¼hrt wurde, kÃ¶nnen die Daten durch folgende Gleichung in den neuen Raum transformiert werden:

$$ X_{PCA} = X_{zentriert} \cdot W $$


Dabei ist:

- $ X_{PCA} $ die Matrix der Daten im reduzierten Raum (mit den Hauptkomponenten),
- $ X_{zentriert} $ die zentrierte Datenmatrix (Originaldaten, von denen der Mittelwert subtrahiert wurde),
- $ W $ die Matrix der ausgewÃ¤hlten Eigenvektoren (Hauptachsen).

Die RÃ¼cktransformation wird durchgefÃ¼hrt, indem man die transformierten Daten wieder in den ursprÃ¼nglichen Merkmalsraum zurÃ¼ckfÃ¼hrt:


$$ X_{rekonstruiert} = X_{PCA} \cdot W^T + \mu $$


Dabei ist:

- $ X_{rekonstruiert} $ die rekonstruierte Datenmatrix,
- $ W^T $ die Transponierte der Eigenvektormatrix,
- $ \mu $ der Mittelwert der ursprÃ¼nglichen Daten (um die zentrierten Daten zurÃ¼ck in den ursprÃ¼nglichen Raum zu transformieren).

#### 2. Warum funktioniert die RÃ¼cktransformation?

Die RÃ¼cktransformation funktioniert, weil die Hauptkomponenten (Eigenvektoren) eine orthogonale Basis fÃ¼r den Merkmalsraum bilden. Die Daten werden zunÃ¤chst in diesen neuen Basisraum transformiert, indem sie entlang der Hauptachsen projiziert werden. Durch die RÃ¼cktransformation kehren wir diese Projektion um, indem wir die Daten wieder in den ursprÃ¼nglichen Merkmalsraum projizieren.

Der Vorteil der RÃ¼cktransformation ist, dass wir sehen kÃ¶nnen, wie gut die ersten $k$ Hauptkomponenten die Originaldaten rekonstruieren kÃ¶nnen. Wenn die Dimensionen stark reduziert wurden, kann die Rekonstruktion jedoch ungenau sein, was durch den Informationsverlust bei der Reduktion bedingt ist.

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
scaled_data
```

```python
x_pca
```

```python
# Schritt 1: Mittelwert berechnen
X_mean = np.mean(scaled_data, axis=0)

# Schritt 2: Manuelle RÃ¼cktransformation
W = pca.components_  # Verwende alle Hauptkomponenten


# Rekonstruktion
X_reconstructed = np.dot(x_pca, W) + X_mean

# ÃœberprÃ¼fen der Dimensionen
print("UrsprÃ¼ngliche Dimensionen:", scaled_data.shape)
print("Rekonstruierte Dimensionen:", X_reconstructed.shape)

# Optional: Vergleiche Original und Rekonstruktion fÃ¼r eine Stichprobe
print("Original Sample:", scaled_data[0])
print("Rekonstruierte Sample:", X_reconstructed[0])

# Ausgabe des rekonstruierten Wertes
reconstructed_sample = pca.inverse_transform(X_pca_2d[0])
print("Rekonstruktion mit sklearn:", reconstructed_sample)
```

```python

# Create a figure
plt.figure(figsize=(8, 6))

# Scatter plot with colors based on the 'target' column
scatter = plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer['target'], cmap='coolwarm', alpha=0.7)

# Adding labels to the axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')

# Adding a legend
plt.legend(handles=scatter.legend_elements()[0], labels=['No Cancer', 'Cancer'])

# Show the plot
plt.show()
```

```python
components = pca.components_

df_components = pd.DataFrame(components, columns=cancer['feature_names'])

print("Contribution of features to the first principal component:")
print(df_components.iloc[0].sort_values(ascending=False))

print("\nContribution of features to the second principal component:")
print(df_components.iloc[1].sort_values(ascending=False))
```

### VerstÃ¤ndnis der BeitrÃ¤ge

Jede Hauptkomponente kann als lineare Kombination der ursprÃ¼nglichen Merkmale ausgedrÃ¼ckt werden:

$$
\text{PC}_1 = w_1 \cdot \text{Merkmal}_1 + w_2 \cdot \text{Merkmal}_2 + \dots + w_n \cdot \text{Merkmal}_n
$$

Hierbei gilt:

- $ \text{PC}_1\ $ ist die erste Hauptkomponente.
- $ w_1, w_2, \dots, w_n\ $ sind die Gewichte, die den einzelnen Merkmalen entsprechen.
- $ \text{Merkmal}_1, \text{Merkmal}_2, \dots, \text{Merkmal}_n\ $ sind die ursprÃ¼nglichen Merkmale.

Zum Beispiel kann die erste Hauptkomponente wie folgt angenÃ¤hert werden:

$$
\text{PC}_1 = 0.260854 \cdot \text{mean concave points} + 0.258400 \cdot \text{mean concavity} + \dots + 0.014531 \cdot \text{smoothness error}
$$


### Interpretation

- **HÃ¶here Faktorwerte**: Merkmale mit hÃ¶heren absoluten Gewichten tragen stÃ¤rker zur Hauptkomponente bei. Zum Beispiel haben â€žmittlere konkave Punkteâ€œ und â€žmittlere KonkavitÃ¤tâ€œ die hÃ¶chsten BeitrÃ¤ge zur ersten Hauptkomponente, was darauf hinweist, dass sie den grÃ¶ÃŸten Einfluss auf die Richtung dieser Komponente haben.
  
- **Vorzeichen der Faktorwerte**: Das Vorzeichen der Ladung zeigt die Richtung des Einflusses an. Eine positive Ladung bedeutet, dass mit steigendem Wert des Merkmals auch der Wert der Hauptkomponente steigt (und umgekehrt bei negativen Ladungen).
