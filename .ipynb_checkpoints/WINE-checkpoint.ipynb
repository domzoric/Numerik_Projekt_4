```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# WINE-Datensatz laden
wine = load_wine()
X = wine.data
y = wine.target
X
y
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df.head()
```

```python
# 1. Zentrieren des Datensatzes
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean
```

```python
# 2. Berechnung der Kovarianzmatrix
cov_matrix = np.cov(X_centered, rowvar=False)
```

```python
# 3. Eigenwertzerlegung der Kovarianzmatrix
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
```

```python
# Die Eigenwerte und Eigenvektoren sortieren
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]
```

```python
# 4. Transformation der Daten
X_pca_manual = np.dot(X_centered, sorted_eigenvectors)
```

```python
# Erste 5 Zeilen der transformierten Daten anzeigen
print(X_pca_manual[:5])
```

```python
# Reduzierung auf die ersten 2 Hauptkomponenten
X_pca_2d = X_pca_manual[:, :2]

# Erste 5 Zeilen der 2D projizierten Daten anzeigen
print(X_pca_2d[:5])
```

```python
import matplotlib.pyplot as plt

# 2D-Plot der ersten 2 Hauptkomponenten
plt.figure(figsize=(10, 8))

# Streudiagramm erstellen, wobei die Farbe der Punkte die Klasse (y) angibt
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap='viridis', s=50)

# Achsen beschriften
plt.xlabel('1st Principal Component')
plt.ylabel('2nd Principal Component')

# Farblegende hinzufgen
plt.colorbar(label='Classes')
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# PCA durchfhren
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Ergebnisse visualisieren
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=40)
plt.xlabel('Erste Hauptkomponente')
plt.ylabel('Zweite Hauptkomponente')
plt.title('PCA mit dem WINE-Datensatz')
plt.colorbar()
plt.show()
```
